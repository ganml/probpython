```python
import random

# Number of simulations
n = 100000

# Counters
count_B = 0
count_A_and_B = 0

for _ in range(n):
    roll = random.randint(1, 6)  # roll a die
    if roll > 3:
        count_B += 1
        if roll % 2 == 0:
            count_A_and_B += 1

P_A_given_B_sim = count_A_and_B / count_B
print(f"Simulated P(A | B): {P_A_given_B_sim:.6f}")
print(f"Difference to theoretical result: {P_A_given_B_sim-2/3:.6f}")
```

```python
import random

n = 100000
positive_tests = 0
disease_and_positive = 0

for _ in range(n):
    has_disease = random.random() < 0.01
    if has_disease:
        test_positive = random.random() < 0.99
    else:
        test_positive = random.random() < 0.05

    if test_positive:
        positive_tests += 1
        if has_disease:
            disease_and_positive += 1

# Empirical posterior probability
P_D_given_T_sim = disease_and_positive / positive_tests
print(f"Simulated P(D | T+): {P_D_given_T_sim:.6f}")
```

```python
import numpy as np

# Number of simulations
N = 100000

# Simulate tossing two fair coins N times
# 0 = Tail, 1 = Head
coin_tosses = np.random.randint(0, 2, size=(N, 2))

# Define events
A = coin_tosses[:, 0] == 1               # First toss is Heads
B = coin_tosses[:, 1] == 1               # Second toss is Heads
C = coin_tosses[:, 0] == coin_tosses[:, 1]  # Both tosses are the same

# Compute probabilities
P_A = np.mean(A)
P_B = np.mean(B)
P_C = np.mean(C)

P_A_B = np.mean(A & B)
P_A_C = np.mean(A & C)
P_B_C = np.mean(B & C)
P_A_B_C = np.mean(A & B & C)

# Display results
print(f"P(A) = {P_A:.4f}, P(B) = {P_B:.4f}, P(C) = {P_C:.4f}")
print(f"P(A ∩ B) = {P_A_B:.4f}, P(A ∩ C) = {P_A_C:.4f}, P(B ∩ C) = {P_B_C:.4f}")
print(f"P(A ∩ B ∩ C) = {P_A_B_C:.4f}")

# Check pairwise independence
print("\nPairwise independence checks:")
print(f"P(A)P(B) = {P_A*P_B:.4f}, P(A ∩ B) = {P_A_B:.4f}")
print(f"P(A)P(C) = {P_A*P_C:.4f}, P(A ∩ C) = {P_A_C:.4f}")
print(f"P(B)P(C) = {P_B*P_C:.4f}, P(B ∩ C) = {P_B_C:.4f}")

# Check mutual independence
print("\nMutual independence check:")
print(f"P(A)P(B)P(C) = {P_A*P_B*P_C:.4f}, P(A ∩ B ∩ C) = {P_A_B_C:.4f}")
```
