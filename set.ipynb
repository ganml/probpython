```python
A = {1,3,4}
B = {4,5}
union_AB = A.union(B)
print(f"Union of A and B: {union_AB}")
intersection_AB = A.intersection(B)
print(f"Intersection of A and B: {intersection_AB}")
difference_AB = A.difference(B)
print(f"Difference of A and B (A - B): {difference_AB}")
difference_BA = B.difference(A)
print(f"Difference of B and A (B - A): {difference_BA}")
symmetric_difference_AB = A.symmetric_difference(B)
print(f"Symmetric Difference of A and B: {symmetric_difference_AB}")
is_subset = B.issubset(A)
print(f"Is B a subset of A? {is_subset}")
is_empty = not B # A set is considered True if it's not empty, False if empty
print(f"Is B empty? {is_empty}")
is_element = 5 in A
print(f"Is 5 an element of A? {is_element}")
```

```python
# Define three example sets
A = {1, 2, 3, 4}
B = {3, 4, 5}
C = {4, 5, 6}

# Distributive law 1: Intersection over Union
left1 = A & (B | C)           # A ∩ (B ∪ C)
right1 = (A & B) | (A & C)    # (A ∩ B) ∪ (A ∩ C)
print("Distributive Law 1:")
print("Left  =", left1)
print("Right =", right1)
print("Law holds:", left1 == right1)
print()

# Distributive law 2: Union over Intersection
left2 = A | (B & C)           # A ∪ (B ∩ C)
right2 = (A | B) & (A | C)    # (A ∪ B) ∩ (A ∪ C)
print("Distributive Law 2:")
print("Left  =", left2)
print("Right =", right2)
print("Law holds:", left2 == right2)
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn3

# Increase global font size for plots
plt.rcParams.update({'font.size': 14})

A = {0, 2, 4}
B = {3, 4, 5}
C = {0, 1, 2}

venn2([A, B], set_labels=('A', 'B'), set_colors=('gray', 'gray'))
plt.title("Venn Diagram of Sets A and B")
plt.savefig("venn2.pdf", bbox_inches="tight")
plt.show()

venn3([A, B, C], set_labels=('A', 'B', 'C'), set_colors=('gray', 'gray', 'gray'))
plt.title("Venn Diagram of Sets A, B, and C")
plt.savefig("venn3.pdf", bbox_inches="tight")
plt.show()
```
