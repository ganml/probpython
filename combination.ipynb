```python
# Addition principle
desserts = ["cake", "pie"]
drinks = ["tea", "coffee", "juice"]

one_item = desserts + drinks
print("One item choices:", one_item)
print("Total:", len(one_item))

# Multiplication principle
meals = [(d, r) for d in desserts for r in drinks]
print("Meal combinations:", meals)
print("Total:", len(meals))
```

```python
import itertools

items = {'A', 'B', 'C', 'D'}
r = 2

combos = list(itertools.combinations(items, r))

print("Combinations:")
for c in combos:
    print(c)

print("Number of combinations:", len(combos))
```

```python
import itertools

# A set of distinct objects
items = {'A', 'B', 'C'}

# Generate all permutations of length 2
perms = list(itertools.permutations(items, 2))

print("All permutations of ['A','B','C']:")
for p in perms:
    print(p)

print("Number of permutations:", len(perms))
```

```python
# Find positive integer solutions to
# x1 + x2 + x3 = 7
# with constraints x1 >= 1, x2 >= 1, x3 >= 1

# Initialize count
count = 0

# Loop through all possible values of x1, x2, x3 satisfying the constraints
for x1 in range(1, 7):       # x1 >= 2
    for x2 in range(1, 7):   # x2 >= 1
        x3 = 7 - x1 - x2
        if x3 >= 1:           # x3 >= 3
            count += 1
            print(f"x1={x1}, x2={x2}, x3={x3}")

print(f"Total number of solutions: {count}")
```
